```python
import pandas as pd
```

```python
df = pd.read_csv("launch_data.csv")
```

```python
df
```

```python
df_groupby = df[["launchyear", "countryoflaunch", "launchmonth"]].groupby(["launchyear", "countryoflaunch"])
df_groupby_count = df_groupby.count().sort_values(["launchyear", "launchmonth"])
df_groupby_count
```

```python
df_groupby_count = df_groupby_count.reset_index()
df_groupby_count
```

```python
df_groupby_count["rank"] = df_groupby_count.groupby("launchyear")["launchmonth"].rank().astype(int)
df_groupby_count
```

```python
df_groupby_count = df_groupby_count.sort_values(["countryoflaunch", "launchyear"])
# df_groupby_count['lastValue'] =
df_groupby_count['lastValue'] = 0
df_groupby_count.to_csv("launch_data_bar_chart_race.csv")
df_groupby_count
```

```python
previous_country, current_country = "", ""
previous_year, current_year, previous_value, current_value = 0, 0, 0, 0
for index, row in df_groupby_count.iloc[0:10, ].iterrows():
    previous_country = current_country
    previous_year = current_year
    previous_value = current_value
    current_country = row["countryoflaunch"]
    current_year = row["launchyear"]
    current_value = row["launchmonth"]
    print(current_country, current_year, current_value, previous_country, previous_year, previous_value)
    if (current_year - previous_year == 1) and (current_country == previous_country) and (previous_value != current_value):
        row['lastValue'] = previous_value
        print(f"Updated lastValue for {current_year} {current_country} to {previous_value}")
df_groupby_count
```
